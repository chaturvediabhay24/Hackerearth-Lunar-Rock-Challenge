```
import tensorflow as tf
from tensorflow.python.keras.layers import Input, Dense, Flatten, MaxPooling2D, Convolution2D
from tensorflow.keras.models import Sequential
from keras.preprocessing import image
```

```
img_height=150
img_width=150
batch_size=32
nb_epochs=100
```

```
classifier = Sequential()

classifier.add(Convolution2D(32, 3, 3, input_shape = (img_height, img_width, 3), activation = 'relu'))

classifier.add(MaxPooling2D(pool_size = (2, 2)))

classifier.add(Convolution2D(32, 3, 3, activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))

classifier.add(Flatten())

classifier.add(Dense(units=32, activation = 'relu'))
classifier.add(Dense(units=1, activation = 'sigmoid'))

classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.summary()
```

```
classifier.load_weights('cp-0065.ckpt')
```

```
import numpy as np
import matplotlib.pyplot as plt
def prediction(link, tar_size, threshold):
  img = image.load_img(link,target_size=tar_size)
  img = image.img_to_array(img)
  img = np.expand_dims(img, axis=0)
  img = img/255
  i=classifier.predict(np.array(img))
  if(i>threshold):
    return "Small"
  else:
    return "Large"
  
```

```
prediction('Train Images/Small/ground0009.png',tar_size=(150, 150, 3), threshold=0.5)
```
